<h3> 3 _ t <br> </h3>
ex) 100 -> (100) = 10 <br>
1, 2, 4, 5, 10, 20, 25, 50, 100 <br>


```python
num = int(input())
sqrt_num = int(num ** (1/2))

front_divisor_list = []
rear_divisor_list = []

for i in range(1, sqrt_num+1):
    if num%i == 0:
        front_divisor_list.append(i)
        if i != int(num/i):
            rear_divisor_list.append(int(num/i))

print(front_divisor_list)
print(rear_divisor_list)
print(front_divisor_list + rear_divisor_list[::-1])
```
