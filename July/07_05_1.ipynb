```python
import numpy as np

num = int(input())
input_arr = []
output_arr = []

for x in range(num):
    input_arr.append(int(input()))

    if len(input_arr)==1:
        output_arr.append(input_arr[0])
    else:
        input_arr.sort()
        length = len(input_arr)
        if  length%2 == 0 :
            output_arr.append(input_arr[length//2 - 1])
        else:
            output_arr.append(input_arr[length//2])

print('============')
for i in output_arr:
    print(i)
```
