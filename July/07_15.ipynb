<h3> 풀이1 _ 함수화

```python
N = int(input())
Que = []

for i in range(N) :
    A = input().split()

    if A[0] == 'push': 
        Que.append(A[1])
        
    elif A[0] == 'pop': 
        if Que: 
            print(Que.pop(0))
        else: 
            print(-1)

    elif A[0] == 'size': 
        print(len(Que))

    elif A[0] == 'empty':
        if len(Que) == 0: 
            print(1)
        else: 
            print(0)
            
    elif A[0] == 'front':
        if len(Que) == 0: 
            print(-1)
        else: 
            print(Que[0])
    
    elif A[0] == 'back':
        if len(Que) == 0: 
            print(-1)
        else: 
            print(Que[-1])
```
