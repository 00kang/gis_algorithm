<h3> 1 _ mine

```python
arr = list(map(int, input().split()))
print(arr)
min = 0
max = 0

for x in range(len(arr)):
    for y in range(x+1,len(arr)):
        if (arr[x] > arr[y]):
            temp = arr[x]
            arr[x] = arr[y]
            arr[y] = temp

max = arr[-1]
min = arr[0]
print(f'max:{max}, min:{min}')
```

```python
num_list = list(map(int, input().split()))

max_num = num_list[0]
min_num = num_list[0]

for num in num_list:
    if max_num < num:
        max_num = num
    if min_num > num:
        min_num = num

print(max_num)
print(min_num)
```
