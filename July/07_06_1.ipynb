```python
num = int(input())
stack_arr = []

for i in range(num):
    command= input().split()
    
    if command[0] == 'push':
        stack_arr.append(command[1])

    elif command[0] == 'pop':
        if len(stack_arr) >= 1:
            print('print: ', stack_arr[-1])
            stack_arr.pop()
        else:
            print('print: -1')

    elif command[0] == 'size':
        print('print: ', len(stack_arr))

    elif command[0] == 'empty':
        if len(stack_arr) >= 1:
            print('print: 0')
        else:
            print('print: 1')
    
    elif command[0] == 'top':
        if len(stack_arr) >= 1:
            print('print: ', stack_arr[-1])
        else:
            print('print: -1')
```
