```python
a, b, R = input().split()
a = int(a)
b = int(b)
R = int(R)

N = int(input())

arr = []

for i in range(N):
  x_i, y_i = input().split()
  x_i = int(x_i)
  y_i = int(y_i)

  if (((x_i-a)**2 + (y_i-b)**2) > R**2):
    arr.append('silent')
  else:
    arr.append('noisy')

print()
for i in arr:
  print(i)
#print(arr)
```

```python
def check(a,b,r,x,y):
    if ((x-a)**2 + (y-b)**2) >= r**2:
        print("silent")
    else:
        print("noisy")

a, b, r = map(int, input().split())
n = int(input())

for _ in range(n):
    x, y = map(int, input().split())
    check(a,b,r,x,y)
```
